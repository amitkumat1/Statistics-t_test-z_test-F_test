```python
print(sample_mean)
```

```python
t_stat, p_value = ttest_1samp(sample['Purchase'], pos_mean)
print(t_stat, p_value)
```

When p_value is less than 0.05, null value is rejected. However in this case p_value is greater so null value will be accepted.

b. It was observed that the percentage of women of the age that spend more than
10000 was 35%. Is it still the same?

### One Sample Test for Proportion

Let's assume, the percentage of women of the age that spend more than 10000 was 35%. Is it still the same?

```python
#null hypothesis - proportion is 35%.
#alternate hypothesis - proportion is not 35%.
```

```python
df_new = df.loc[(df['Purchase']>10000)]
```

```python
df_new.shape
```

```python
df_new['Gender'].value_counts()
```

```python
count = df_new['Gender'].value_counts()[0]
```

```python
count
```

```python
new = len(df_new['Gender'])
```

```python
new
```

```python
p0 = 0.35
```

```python
df_new['Gender'].value_counts()[0]/new
```

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
z_stat, p_value = proportions_ztest(count = count,
                                   nobs = new,
                                   value = p0,
                                   alternative = 'two-sided',
                                   prop_var = False)
print(z_stat, p_value)
```

Since p_value is less than 0.05 hence, we have to reject null hypothesis.So, the percentage of women of the age that spend more than 10000 was not 35%.

c. Is the average purchase made by men and women of the age 18-25 same?

### Two Sample Test for mean

let's assume, Is the average purchase made by men and women of the age 18-25 same?

```python
#null hypothesis - average spends are equal
#alternate hypothesis - average spends are not equal

df_men = df.loc[(df['Gender'] == 1)& (df['Age'] == 1)]
df_women = df.loc[(df['Gender'] == 0) & (df['Age'] == 1)]
```

```python
df_men_sample = df_men.sample(500, random_state=0)
df_women_sample = df_women.sample(500, random_state=0)
```

```python
print(df_men_sample.Purchase.var())
print(df_women_sample.Purchase.var())
```

```python
print(df_men_sample.Purchase.mean())
print(df_women_sample.Purchase.mean())
```

```python
from scipy.stats import ttest_ind

t_stat_2, p_val_2 = ttest_ind(df_men_sample.Purchase, df_women_sample.Purchase, equal_var=False)
print(t_stat_2, p_val_2)
```

Since, p_value is less than 0.05 hence, we can reject null hypothesis. So, purchase made by men and women of the age 18-25
is not same

d. Is the percentage of men who have spent more than 10000 the same for the
ages 18-25 and 26-35?

### Two Sample test for Proportion
lets assume, Is the percentage of men who have spend more than 10000 same for the ages 18-25 and 26-35

```python
#null hypothesis - Percentage of the men in the age groups is same
#alternate hypothesis - Percentage of the men in the age groups is not same

df_age1 = df.loc[(df['Age'] == 1) & (df['Purchase'] > 10000)]
df_age2 = df.loc[(df['Age'] == 2) & (df['Purchase'] > 10000)]
```

```python
len(df_age1), len(df_age2)
```

```python
df_age1_sample = df_age1 .sample(1000, random_state=0)
df_age2_sample = df_age2.sample(1000, random_state=0)
```

```python
count = [(df_age1_sample['Gender'] == 1).sum(), (df_age2_sample['Gender'] == 1).sum()]

nobs = [(len(df_age1_sample)), len(df_age2_sample)]
```

```python
from statsmodels.stats.proportion import proportions_ztest
stat_2sample, p_value_2sample = proportions_ztest(count=count,
                                                  nobs=nobs,
                                                  value=0,
                                                  alternative='two-sided',
                                                  prop_var=False)
```

```python
print(stat_2sample, p_value_2sample)
```

Since, p value is more than 0.05, we cannot reject the null hypthesis.
Hence, Null hypothesis is accepted. Therefore, Percentage of the men in  both the age groups who have spent more than 10000 is same
